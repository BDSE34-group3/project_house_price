```python
import pandas as pd
```

```python
import lib.corcoordinate_caculate as cc
```

```python
near_fac = pd.read_excel('../../all_data/比鄰設施_final表.xlsx')
```

```python
# latitude緯度，longitude經度
# Longitude, Latitude錯置整理(白癡寫法)

# 把錯置的取出並轉正常
near_fac_fix = near_fac[ near_fac['Longitude'] > 100]
near_fac_fix.columns
near_fac_fix = near_fac_fix.rename(columns={'機關名稱':'機關名稱', '機關地址':'機關地址', 'Longitude':'Latitude', 'Latitude':'Longitude', '類型':'類型', 'cor':'cor'})
near_fac_fix = near_fac_fix.reindex(columns=['機關名稱', '機關地址', 'Longitude', 'Latitude', '類型'])

# 原先正常的
near_fac_normal = near_fac[ near_fac['Longitude'] < 100]

# 最後再合起來
near_fac = pd.concat([near_fac_normal, near_fac_fix])
```

```python
cc.convert_lat_lon_to_tuple(near_fac, 'Longitude', 'Latitude')
```
