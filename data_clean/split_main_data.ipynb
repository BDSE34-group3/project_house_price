* 若切成每個都是獨立檔案的話，就可以跑multi processing了ㄟ...

```python
import pandas as pd
import os
```

```python
house_main_df = pd.read_csv('../../all_data/主資料集/台北市_10101_11303_房價data.csv')
```

```python
folderPath = "./main_data_piece_1000/"
if not os.path.exists(folderPath):
    os.makedirs(folderPath)

# 將每1000筆資料切成一個獨立的檔案
for number_row in range(1000, 244001, 1000): #244001
        house_df = house_main_df.iloc[number_row - 1000 : number_row, :].copy()
        house_df.to_csv(f"./main_data_piece_1000/main_data_{number_row - 1000}_{number_row}.csv")
```
