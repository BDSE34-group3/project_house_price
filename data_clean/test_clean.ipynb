```python
# 資料表格都相同，所以寫成函式做轉換
def convert_table(table):
    new_t = table.pivot_table('數值', index='年月', columns='名稱')
    return new_t
# 代號名稱對應表
def mapping_list(table):
    mapping_list_table = table[['名稱', '代號']]
    uni_set = set()
    # dataframe跟dict很像，要.values才會是值，不然預設會取到column的名稱(相對於dict也就是key的概念)
    for index, value in enumerate(mapping_list_table.values):
        uni_set.add((mapping_list_table['代號'].iloc[index], mapping_list_table['名稱'].iloc[index]))
    return list(uni_set)
```

```python
arch_stas_pivot = convert_table(arch_stas)
arch_stas_list = mapping_list(arch_stas)
```

```python
concrete_stock = pd.read_excel('../all_data/經濟指標/資料來源_XQ/月頻率資料/201101_202405_水泥業_月.xlsx')

concrete_stock['水泥(TSE11)'] = pd.to_datetime(concrete_stock['水泥(TSE11)'])

# 新增一個月頻率的欄位，用來合併經濟指標用的
concrete_stock['year_month_concrete_stock'] = concrete_stock['水泥(TSE11)'].dt.to_period('M')
concrete_stock.dropna(axis=1, inplace=True)
```

```python
concrete_stock
```

```python
ship_stock = pd.read_excel('../all_data/經濟指標/資料來源_XQ/月頻率資料/201101_202405_航運業_月.xlsx')
ship_stock['航運業(TSE26)'] = pd.to_datetime(ship_stock['航運業(TSE26)'])

# 新增一個月頻率的欄位，用來合併經濟指標用的
ship_stock['year_month_ship_stock'] = ship_stock['航運業(TSE26)'].dt.to_period('M')
ship_stock.dropna(axis=1, inplace=True)
```

```python
ship_stock
```

```python
iron_stock = pd.read_excel('../all_data/經濟指標/資料來源_XQ/月頻率資料/201101_202405_鋼鐵業_月.xlsx')
iron_stock['鋼鐵(TSE20)'] = pd.to_datetime(iron_stock['鋼鐵(TSE20)'])

# 新增一個月頻率的欄位，用來合併經濟指標用的
iron_stock['year_month_iron_stock'] = iron_stock['鋼鐵(TSE20)'].dt.to_period('M')
iron_stock.dropna(axis=1, inplace=True)
```

```python
iron_stock
```

```python
arch_stock = pd.read_excel('../all_data/經濟指標/資料來源_XQ/月頻率資料/201101_202405_營造業_月.xlsx')
arch_stock['建材營造(TSE25)'] = pd.to_datetime(arch_stock['建材營造(TSE25)'])

# 新增一個月頻率的欄位，用來合併經濟指標用的
arch_stock['year_month_arch_stock'] = arch_stock['建材營造(TSE25)'].dt.to_period('M')
arch_stock.dropna(axis=1, inplace=True)
```

```python
arch_stock
```

```python
# suffixes=('_arch', '_iron') 若合併的df有同名的欄位，要加什麼贅詞用以區分
mix = pd.merge(arch_stock, iron_stock, left_on='year_month_arch_stock', right_on='year_month_iron_stock', how='inner', suffixes=('_arch', '_iron'))
```

```python
mix2 = pd.merge(concrete_stock, ship_stock, left_on='year_month_concrete_stock', right_on='year_month_ship_stock', suffixes=('_concrete','_ship'))
```

```python
mix2.info()
```

```python
mix_all = pd.merge(mix, mix2, left_on='year_month_arch_stock', right_on='year_month_concrete_stock')
```

```python
# 資料來源XQ的資料合併集
mix_all.head()
```

```python
f1 = pd.merge(df,mix_all, left_on='year_month_for_combine', right_on='year_month_arch_stock')
```

```python
mm_all
```

```python
# 要把mm_all的index新增成一個欄位，並把他的資料型態轉為period[M]，用以與主資料集進行合併
append_col = mm_all.index
append_col = pd.Series(append_col, index=mm_all.index)
mm_all['date'] = append_col
mm_all['date'] =  pd.to_datetime(mm_all['date'])
mm_all['date'] = mm_all['date'].dt.to_period('M')
```

```python
final_df = pd.merge(f1, mm_all, left_on='year_month_for_combine', right_on='date', how='inner')
```

```python
# 最終合併的資料表
final_df
```
