```python
import pandas as pd
```

```python
df = pd.read_excel('../../all_data/房價指數/201101_202405_信義房屋指數.xlsx')
```

```python
df['年份'] = df['年份'].sort_values(ascending=False)
```

```python
df
```

```python
df['date'] = pd.to_datetime(df['年份'].astype(str) + '-' + df['月'].astype(str), format='%Y-%m')
```

```python
df['date'] = df['date'].dt.to_period('M')
df.sort_values('date', inplace=True)
```

```python
df.drop([0,1],axis=0, inplace=True)
df.drop(['年份', '月'],axis=1, inplace=True)
```

```python
df.columns
```

```python
df.columns = ['house_index_xyi_tp', 'house_index_xyi_new_tp', 'house_index_xyi_all_tp', 'date']
```

```python
df.to_csv('../../all_data/房價指數/xyi_house_index.csv', encoding='utf-8', index=False)
```
