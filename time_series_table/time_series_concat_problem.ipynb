## 把每一筆時間的，特徵資料變成二維的
* 原先 X .shape = (180,) -> 更新後 X_new.shape = (5, 36)

```python
# 有喔~老師有說過怎麼把一條(1D)變成成2D
# 都先從最小的軸開始做axis=0
# 所以若原先形狀為 X = [x_1_lag_1   x_1_lag_2  x_2_lag_1  x_2_lag_2   x_3_lag_1  x_3_lag_2]

# 第一步axis=0代表的是timestamp(此次中有2個)所以形狀就會被切成，axis=0切兩分的意思就是把把原先的一長條切成兩分(等分)，這樣axis=0軸的數量就會等於2

## 這邊形狀寫錯了(寫反了) 應該是
'''
# [[x_1_lag_1   x_2_lag_1  x_3_lag_1],  
   [x_1_lag_2   x_2_lag_2  x_3_lag_2]]

'''


####
'''
# [[x_1_lag_1   x_1_lag_2  x_2_lag_1],  
   [x_2_lag_2   x_3_lag_1  x_3_lag_2]]
'''
# 將特徵轉換為所需的形狀 (2, 3)
# new_X = X.reshape(2, 3)
```

```python
one_data = vars_shift_df.drop('date', axis=1)
one_data = one_data.iloc[0]

# # target_y就是預測值
y = one_data['target_index_house_index_xyi_all_tp']
# # 所以對於多變量時間序列特徵而言
X = one_data.drop('target_index_house_index_xyi_all_tp')
# # 此時X的shpae應該為 
print(f"X shape:{X.shape}") # 180是有16個特徵，每個有5個lag

new_X = X.values.reshape(5,36)
print(f"new_X shape:{new_X.shape}")
print('='*50)

print(f"{X}")

print(f"{new_X}")
```
